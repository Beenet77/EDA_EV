### Handling missing values 

```python
df.head()
```

```python
categorical_columns = ['Model', 'Vehicle Location', 'Electric Utility']
categorical_imputer = SimpleImputer(strategy='most_frequent')
df[['Model']] = categorical_imputer.fit_transform(df[['Model']])
df[['Vehicle Location']] = categorical_imputer.fit_transform(df[['Vehicle Location']])
df[['Electric Utility']] = categorical_imputer.fit_transform(df[['Electric Utility']])
```

```python
numeric_imputer = SimpleImputer(strategy='median')
df['Legislative District'] = numeric_imputer.fit_transform(df[['Legislative District']])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
## Outliers detection
numeric_columns = df.select_dtypes(include=[np.number]).columns
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return ((data < lower_bound) | (data > upper_bound)).sum()

print("Outliers detected using IQR method:")
for col in numeric_columns:
    outliers = detect_outliers_iqr(df[col])
    print(f"{col}: {outliers} outliers")
```

```python

for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='blue')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

3. The steady increase in electric range over time reflects improvements in battery technology and vehicle efficiency.
4. Early electric vehicles (pre-2010) had lower ranges, though a few outliers show above-average performance for their time.

### Task 1 : Create a Choropleth using plotly.express to display the number of EV vehicles based on location.


```python
# Aggregate data by state
state_counts = df['State'].value_counts().reset_index()
state_counts.columns = ['State', 'Count']

# Create the Choropleth map
fig = px.choropleth(state_counts,
                    locations='State', 
                    locationmode="USA-states", 
                    color='Count',
                    scope="usa",
                    color_continuous_scale="Viridis",
                    range_color=(0, 10000),
                    title="Number of Electric Vehicles by State")

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()

# Print top 5 and bottom 5 states by EV count
print("Top 5 States by EV Count:")
print(state_counts.head())
print("\nBottom 5 States by EV Count:")
print(state_counts.tail())
```

#### Insight 
1. WA state has the highest number of EV's i.e:11k and the state like MS,AL,DE,OK,ND have got the lowest number of EV's.

### Task 2: Create a Racing Bar Plot to display the animation of EV Make and its count each year.

```python
import bar_chart_race as bcr
```

```python
ev_brand_by_year = df.groupby(['Model Year', 'Make']).size().reset_index(name='EV Count')

# Step 2: Create a list of all unique makes
unique_makes = df['Make'].unique()

# Step 3: Ensure all makes appear in every year by filling missing combinations
all_years = pd.DataFrame({'Model Year': sorted(df['Model Year'].unique())})
all_combinations = all_years.assign(key=1).merge(pd.DataFrame({'Make': unique_makes, 'key':1}), on='key').drop('key', axis=1)
ev_brand_by_year_full = all_combinations.merge(ev_brand_by_year, on=['Model Year', 'Make'], how='left').fillna(0)

# StepConvert EV Count to integer (since it was NaN before)
ev_brand_by_year_full['EV Count'] = ev_brand_by_year_full['EV Count'].astype(int)

#  Creating the animated racing bar plot 
fig = px.bar(
    ev_brand_by_year_full,  # Data
    x='EV Count',
    y='Make', 
    color='Make', 
    animation_frame='Model Year',  
    orientation='h',  
    title='Electric Vehicle brands Over the Years',
    labels={'EV Count':'Number of EVs', 'Make':'Car Brands'},  # Axis labels
    range_x=[0, ev_brand_by_year_full['EV Count'].max() * 1.1],  # Dynamically set x-axis range
    height=900  # Increased height for better visibility
)


fig.show()

```
